### Ход работы

#### Постановка задачи

<span style="font-style:normal; font-size:12pt;"> 
<p>
    1. Путем приведения дифференциального уравнения к нормальной системе первого порядка получим:
</p>
    
$$
 \begin{cases}
    \dot{x} = x_{1} \\
    \dot{x_{1}} = \alpha x - \beta x^2 - d x_{1} \ \ (1) \\
    d, \alpha, \beta > 0
  \end{cases}
$$

<span style="font-style:normal; font-size:12pt;"> 
<p>
   2. Чтобы найти стационарные точки, нужно приравнять к нулю правые части уравнений системы:
</p>
    
        
$$
 \begin{cases}
    x_{1} = 0 \\
    \alpha x - \beta x^2 - d x_{1} = 0 \\
    d, \alpha, \beta > 0
  \end{cases}
$$
  
    
<p>
   Корни легко находятся аналитически. Получим две пары стационарных точек:
</p>  
  

$$
\left[ 
    \begin{gathered} 
        \begin{cases}
            x = 0 \\
            x_{1} = 0
        \end{cases}
        \\
        \begin{cases}
            x = \frac{\alpha}{\beta} \\
            x_{1} = 0
        \end{cases}
        \\
        d, \alpha, \beta > 0
    \end{gathered} 
\right.
$$
    

<p>  
    Пусть
</p>
$$
 \ \  \alpha = 1, \ \beta = 0.5, \  d = 2 \\
$$

##### Фазовые портреты

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f1(x, x1, alpha, beta, d):
    return x1    
    
def f2(x, x1, alpha, beta, d):
    return alpha * x - beta * x**2 - d * x1
    
```

```python
alpha = 1
beta  = 0.5
d     = 2
```

```python
x0   = 0
x0_1 = 0

x1   = alpha/beta
x1_1 = 0
```

```python
# hx = [2, 3]
# hy = [0, -1]

# h1x = [0, 1 + 2**(1/2)]
# h1y = [0, 1]

# h2x = [0, 1 - 2**(1/2)]
# h2y = [0, 1]
```

```python
# Построим координатную сетку фазовой плоскости

#границы
a = -4
b = 4

# шаг
step = 0.3

x_plot   = np.linspace(a, b, round((b - a) / step) + 1)
x_1_plot = np.linspace(a, b, round((b - a) / step) + 1)
```

```python
X, Y = np.meshgrid(x_plot, x_1_plot)
```

```python
U = f1(X, Y, alpha, beta, d)
V = f2(X, Y, alpha, beta, d)
```

```python
fig, ax = plt.subplots(figsize=(20, 12))
q = ax.quiver(X, Y, U, V)
ax.quiverkey(q, X=0.3, Y=1.1, U=10,
             label='Quiver key, length = 10', labelpos='E')

ax.scatter([x0], [x0_1], color="red")
ax.scatter([x1], [x1_1], color="red")

# ax.plot(hx,hy, color="blue")

# ax.plot(h1x,h1y, color="blue")
# ax.plot(h2x, h2y, color="blue")


plt.show()
```
